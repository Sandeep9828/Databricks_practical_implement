```python
from pyspark.sql.functions import *

df =  spark.createDataFrame([
    ( [4,2,1,6,4,7,8,6], [5,6,7,3,5,2,1] )
], ["lst1", "lst2"])

new_zip = df.withColumn("new_zip_col",arrays_zip('lst1','lst2') )

new_zip.display()
```

```python
l1 = [4,2,1,6,4,7,8,6]
l2 = [5,6,7,8,3,5,2,1]

print(list(zip(l1,l2)))
```

```python
data = [
    ("John",  [4, 6, 7, 9, 2],   [1, 2, 3, 7, 7]),
    ("David", [7, 5, 1, 4, 7, 1],[3, 2, 8, 9, 4, 9]),
    ("Mike",  [3, 9, 1, 6, 2],   [1, 2, 3, 5, 8])
]

cols = ["Name", "Array_1", "Array_2"]
df = spark.createDataFrame(data, cols)

result_df = df.withColumn("Intersect", array_intersect("Array_1", "Array_2"))
result_df.show(truncate=False)
```

```python
df = df.withColumn("Uncommon",
                   array_except(col("Array_1"), col("Array_2")))
```

```python
df.display()
```

```python
df = df.withColumn("sorted_array", array_sort(col("Array_1")))
```

```python
from pyspark.sql.functions import array_sort, reverse

df2 = df.withColumn("sorted_desc", reverse(array_sort("Array_1")))
```

```python
df2.display()
```
