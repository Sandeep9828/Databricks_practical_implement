```python
from pyspark.sql import functions as F
```

```python
data = [
    (1,7.5),(2,15.0),(3,30.0)
]
df = spark.createDataFrame(data, ['cust_id','amount'])
```

```python
df.display()

```

```python
out = df.select("cust_id","amount",F.when(F.col("amount")<10,"small")\
                            . when( (F.col("amount")>=10) & (F.col("amount")<20),"Medium")\
                            . when( F.col("amount")>= 20 ,'Large' )\
                            .otherwise("Invalid").alias("Type"))

out.display()
```
