```python

from pyspark.sql import functions as f
from pyspark.sql.functions import broadcast

# 1. Big fact table
big_df = spark.range(1, 5_000_000).withColumn("id", (f.col("id") % 1000))
```

```python
dim_df = spark.createDataFrame(
    [(i, f"Category_{i}") for i in range(1000)],
    ["id", "category"]
)
```

```python
dim_df.display()
```

```python
result = big_df.join(dim_df,"id")

print("withOut broadcast join")
result.explain()
```

```python
result2 = big_df.join(broadcast(dim_df), "id")

print("With Broadcast")
result2.explain(True)



'''
Output will show:
	•	❌ SortMergeJoin for non-broadcast
	•	✅ BroadcastHashJoin for broadcast

'''
```
