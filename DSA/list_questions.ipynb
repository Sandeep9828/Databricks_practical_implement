```python
### another Brute solution 

### pop and insert

nums = [5,-2,3,9,0,6,10,7]

n = len(nums)
r = 12
k = r%n

for _ in range(k):
    if _ >len(nums):
        break
    e = nums.pop() #### Time complexity >> O(1)
    nums.insert(0,e) #### Time comlpexity >> O(N)

print(nums)


'''
total complexity >> O(r*n)
'''
```

```python
### Optimal solution using slicing 

nums = [5,-2,3,9,0,6,10,7]
k = 3
n = len(nums)
k = k%n

nums[:] = nums[n-k:]  + nums[:n-k]


print(nums)

'''
Time complexity for first slicing >> O(k) + O(n-k) = O (k+n-k) => O(N)
space complexity >> O(1)
'''
```

```python
### Without Slicing
nums = [5,-2,3,9,0,6,10,7]
k = 3
n = len(nums)


def reverse_s(nums,left,right):
    while left<right:
        nums[left],nums[right] = nums[right],nums[left]
        left +=1
        right -=1

reverse_s(nums,n-k,n-1) #### TC>> O(k/2)
print(nums)

reverse_s(nums,0,n-k-1) #### TC >> O((n-k)/2)
print(nums)

reverse_s(nums,0,n-1)   #### TC >> O(n/2)
print(nums)


'''
Time complexity >> total time complexity >>> O(k/2 + (n-k)/2 + (n-1)/2) ~~ O(n)

space complexity >> O(1)

'''
```
