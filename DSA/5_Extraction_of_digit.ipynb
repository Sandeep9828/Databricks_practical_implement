```python
#### Hashing consept
'''
problem - here i want to find the elements from m present in side the n or not and if present then how many time present i want to figure it out.

'''

'''
Constrains
    1. 1<= n[i] <=10
    2. n can have 10**8
    3. m can have 10**8
    
'''


n = [5,3,2,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]


for i in m:
    count = 0
    for j in n:
        if i ==j:
            count +=1
    print(f"Element {i} -  {count}")


'''
this is the solution, the time complexity of this 

Time Complexity - O(m*n)
Space complexity - O(1)


in worst case if we have total 10**8 element in both the array then the time complexity will be O(10**16)

and here we will get the error TLE because above 10**8 element we will get the error TLE. so we have to optimize the code.

'''
```

```python
#### Optimize solution of above code


'''
Constrains
    1. 1<= n[i] <=10
    2. n can have 10**8
    3. m can have 10**8
    
'''


n = [5,3,2,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]


### so first of we will create hash map like dictionary in which we will store the number of elements and counts

d = {}
for i in range(len(n)):
    d[n[i]] = d.get(n[i],0)+1

for j in m:
    if j <1 or j > 10:
        print(0)
    else:
        print(F"{j} - {d.get(j)}")
# print(d)

"""
so time compexity of this code will - O( N+M ) - how >> for first loop complexity is O(10**8) and for second loop complexity is O(10**8) so total complexity is 2*(10**8) ~ ~ O(10**8)

space complexity -- O(11) because this is mentioned in the constrains that the element will be between 1 to 10 so we will create the dictionary of size 11
"""
```
