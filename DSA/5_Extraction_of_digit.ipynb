```python

n = 121
num = n
result = 0

while num>0:
    ld = num%10
    result = ld + result*10
    num = num//10

if result ==n:
    print("number is palindrome")
else:
    print("not palindrome")


''' 

##### Time complexity #######
here Number of iteration depends on the  "num = num//10" so every time it devide by constant value 10 so time complexity is O(log10 (n)). if the num//5 then time complexity is O(log5(n))



### Space Complexity

so here i am using constant space so space complexity is O(1)

'''
```
