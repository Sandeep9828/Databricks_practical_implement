```python
def factorial(n):
    if n < 2:
        return 1

    return n * factorial(n-1)

print(factorial(4))




'''
Time complexity will be O(n)
space Complexity will be O(n) - stake space
'''
```

```python

n = [5,9,8,3,6,7,1,4,2]

l = 2
r = 5


'''
so question is to reverse the number as given in the range only l to r where l is left and r is right  using recursion.
'''

def reverse(n,l,r):
    if l >=r:
        return 1
    
    n[l] ,n[r] = n[r],n[l]
    reverse(n,l+1,r-1)

    return n
    
print(reverse(n,l,r))
```

```python

n = [5,9,8,3,6,7,1,4,2]
l = 2
r = 5
while l<=r:
    n[l],n[r] = n[r],n[l]
    l +=1
    r -=1

print(n)
```

```python

s = "sandeep"
n= len(s)
l = 0
r = n-1


def palindrome_check(s,l,r):
    while l<r :
        if s[l] != s[r]:
            return False
        l +=1
        r -=1
        return True
    

if palindrome_check(s,l,r) == True:
    print("Palindrome")
else:
    print("Not Palindrome")
'''

time complexity will be O(n)
space complexity will be O(1) ]

'''
```

```python
n =5

def fibonacci(sum,n):
    if n<=0:
        return sum
    
    sum = sum+n
    
    return fibonacci(sum,n-1)

print(fibonacci(0,2))
```
