```python

### selection sort

lst = [1,7,8,3,4,5,6,2]

''' this need to sort '''

n = len(lst)


for i in range(n):
    min_index = i

    for j in range(i+1,n):
        if lst[j] < lst[min_index]:
            min_index = j
        # lst[i],lst[min_index] = lst[min_index],lst[i]
    lst[min_index],lst[i] = lst[i],lst[min_index]

print(lst)



'''
time complexity of this O(n(n+1)/2) ~~ n**2
space complexity -> O(1)

'''
```

```python

nums = [5,8,1,6,9,2,4]
n = len(nums)

for i in range( n-2,-1,-1 ):

    for j in range(0,i+1):
        if nums[j] > nums[j+1]:
            nums[j+1],nums[j] = nums[j], nums[j+1]

print(nums)




'''
time complexity -> O(n(n+1)/2) ~~ O(n**2)
space complexity -> O(1)

'''
```

```python

lst =[3,5,6,4,8,9,10,7,1]
n = len(lst)



for i in range(1,n):

    key = lst[i] ## >> 5
    j = i-1 ### >> 0

    while lst[j]>key and j >=0:

        lst[j+1] = lst[j]
        j= j-1
    
    lst[j+1] = key

print(lst)

'''
time complexity - O(n (n+1)/2) ~~ O(n**2)
space complexity -> O(1)

'''
```
